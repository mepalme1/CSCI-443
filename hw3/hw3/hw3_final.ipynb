```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 6.1
np.random.seed(42)

mu, sigma = 0, 0.5
m, n = 100, 5

# PDF of log-normal
x = np.linspace(0.01, 6, 1000)
pdf = stats.lognorm.pdf(x, s=sigma, scale=np.exp(mu))

# sample means
sample_means = [np.mean(np.random.lognormal(mu, sigma, n)) for _ in range(m)]

# PDF with histogram overlay
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, pdf, 'r-', linewidth=2, label='Log-Normal PDF')
ax.hist(sample_means, bins=20, density=True, alpha=0.6, color='blue', label=f'Sample Means (m=100, n=5)')
ax.set_xlabel('Value')
ax.set_ylabel('Density')
ax.set_title(f'Log-Normal PDF with Histogram of {m} Sample Means (n={n})')
ax.legend()
plt.show()

# sample means on the real line
fig, ax = plt.subplots(figsize=(10, 2))
ax.scatter(sample_means, np.zeros_like(sample_means), s=100, color='blue')
ax.set_xlabel('Sample Mean Value')
ax.set_title(f'100 Sample Means on the Real Line (n=5)')
ax.set_yticks([])
plt.show()
```

```python
# 6.2
m2 = 10000
n2 = 5

sample_means_2 = [np.mean(np.random.lognormal(mu, sigma, n2)) for _ in range(m2)]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, pdf, 'r-', linewidth=2, label='Log-Normal PDF')
ax.hist(sample_means_2, bins=50, density=True, alpha=0.6, color='green', label='Sample Means (m=10000, n=5)')
ax.set_xlabel('Value')
ax.set_ylabel('Density')
ax.set_title('Log-Normal PDF with Histogram of 10000 Sample Means (n=5)')
ax.legend()
plt.show()
```

```python
# 6.3
m3 = 1000
n3 = 10

sample_means_3 = [np.mean(np.random.lognormal(mu, sigma, n3)) for _ in range(m3)]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, pdf, 'r-', linewidth=2, label='Log-Normal PDF')
ax.hist(sample_means_2, bins=50, density=True, alpha=0.6, color='green', label='Sample Means (m=1000, n=10)')
ax.set_xlabel('Value')
ax.set_ylabel('Density')
ax.set_title('Log-Normal PDF with Histogram of 1000 Sample Means (n=10)')
ax.legend()
plt.show()
```
